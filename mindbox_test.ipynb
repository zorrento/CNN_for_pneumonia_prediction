```python
def count_group_entries(n_customers):
  answer_dict = {0 : 1}
  for i in range(1, n_customers):
    temp_sum = 0
    k = i
    while k > 0:
      digit = k % 10
      temp_sum += digit
      k = k // 10
    answer_dict.update({temp_sum : answer_dict.get(temp_sum, 0) + 1})
  return answer_dict #возвращает в качестве словаря
```
